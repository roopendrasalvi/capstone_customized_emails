```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
headers = {"User-Agent": "Guest"}
```

```python
response = requests.get("https://www.linkedin.com/company/yash-technologies/", headers=headers)
```

```python
response.status_code
```

```python
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
title_tag = soup.find('title')
designation_tag = soup.find('h2')
followers_tag = soup.find('meta', {"property": "og:description"})
description_tag = soup.find('p', class_='break-words')
```

```python
name = title_tag.get_text(strip=True).split("|")[0].strip() if title_tag else 'Profile Not Found'
designation = designation_tag.get_text(strip=True) if designation_tag else 'Designation Not Found'
followers_match = re.search(r'\b(\d[\d,.]*)\s+followers\b', followers_tag['content']) if followers_tag else None
followers_count = followers_match.group(1) if followers_match else 'Followers Not Found'
description = description_tag.get_text(strip=True) if description_tag else 'Description Not Found'
```

```python
print("Name:", name)
print("Designation:", designation)
print("Followers:", followers_count)
print("Description:", description)
```
